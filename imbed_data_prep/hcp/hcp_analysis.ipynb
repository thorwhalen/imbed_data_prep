# Making link data

```python
from typing import Sequence, Callable, Union, Optional
from functools import partial
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.base import BaseEstimator

import pandas as pd
from dataclasses import dataclass

dflt_knn_maker = partial(NearestNeighbors, n_neighbors=100, metric='cosine')


def is_vector_of_keys(v):
    first_element = next(iter(v))
    return isinstance(first_element, int)

@dataclass
class KnnSearch:
    vectors: Sequence
    keys: Optional[Sequence] = None
    knn: Union[int, Callable] = dflt_knn_maker

    def __post_init__(self):
        if isinstance(self.knn, int):
            n_neighbors = self.knn
            self.knn = dflt_knn_maker(n_neighbors)
        elif callable(self.knn):
            knn_factory = self.knn
            self.knn = knn_factory()

        if isinstance(self.vectors, pd.Series):
            series = self.vectors
            self.vectors = np.vstack(series.values)
            if self.keys is None:
                self.keys = series.index.values

        assert len(self.vectors) == len(self.keys), (
            "vectors and keys must have the same number of element"
        )
        
    def fit(self):
        self.knn.fit(np.vstack(self.vectors))
        return self

    def vectors_for_keys(self, keys):
        indices = np.where(np.isin(self.keys, keys))[0]
        return self.vectors[indices]
    
    def search(self, query_vectors, *, with_distances=True):
        # if query_vectors is a single vector, reshape it so sklearn kneighbors will accept it
        if is_vector_of_keys(query_vectors):
            query_vectors = self.vectors_for_keys(query_vectors)
            
        if len(query_vectors.shape) == 1:
            query_vectors = query_vectors.reshape(1, -1)
        
        distances, indices = self.knn.kneighbors(query_vectors)
        matched_keys = [self.keys[i] for i in indices]
        if with_distances:
            return distances, matched_keys
        else:
            return matched_keys
```
