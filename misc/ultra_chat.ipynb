```python
from imbed_data_prep.ultra_chat import get_raw_data

data = get_raw_data()
```

```python
t = data['train']
tt = t.data
```

```python
df = tt.to_pandas()
print(f"{df.shape=}")
df.iloc[0]
```

```python
df.data.iloc[0]
```

```python
n_turns = list(map(len, df.data))
# histogram of number of turns
import matplotlib.pyplot as plt

print(f"{sum(n_turns) = }")
plt.hist(n_turns);
plt.title("number of turns per row (session?)")
```

```python
import itertools

text_sizes = list(map(lambda x: map(len, x), df.data))
text_sizes = list(itertools.chain(*text_sizes))
plt.hist(text_sizes, bins=100);
plt.title('text sizes')
```
